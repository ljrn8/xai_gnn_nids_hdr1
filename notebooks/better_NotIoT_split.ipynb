```python
from tqdm import tqdm

chunks = len(df) // 10_000
test_chunks_list = np.random.choice(chunks, size=int(chunks * 0.2), replace=False)
train_chunks_list = [i for i in range(chunks) if i not in test_chunks_list]

test_df = pd.DataFrame()
train_df = pd.DataFrame()

for i in tqdm(range(chunks)):
    chunk = df.iloc[i*10_000:(i+1)*10_000]
    if i in test_chunks_list:
        test_df = pd.concat([test_df, chunk])
    else:
        train_df = pd.concat([train_df, chunk])

```

```python
len(test_df) / (len(train_df) + len(test_df))
```

```python
# attack:mel proportion of both datasets
sum(test_df.Attack != 'Benign' ) / ( sum(test_df.Attack != 'Benign' )  + sum(test_df.Attack == 'Benign') ), sum(train_df.Attack != 'Benign' ) / ( sum(train_df.Attack != 'Benign' ) + sum(train_df.Attack == 'Benign'))
```

```python
# get the noramlize proportion of each attack label for test and train dfs
for a in np.unique(train_df.Attack):
    test_p = sum(test_df.Attack == a) / len(test_df)
    train_p = sum(train_df.Attack == a) / len(train_df)
    print(f'{a} test: {test_p:.5f} | train: {train_p:.5f}')
```

```python
test_df.to_csv('../interm/ToNIoT_processed_test.csv')
print('done test')
train_df.to_csv('../interm/ToNIoT_processed_train.csv')
print('done')
```
