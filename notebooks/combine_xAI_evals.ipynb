```python
import pickle

plt.figure(figsize=(4, 4))

for metric in ["fp", "fm"]:
	for file in (
	 '..\\output\\CICIDS\\npge\\eval\\edge mask sparsity variation.csv',
	 '..\\output\\CICIDS\\lpge\\eval\\edge mask sparsity variation.csv'
	):
		with open(file, "rb") as f:
			sparsities, fps, fms, threshes = pickle.load(f)
		
		if metric == "fp": 
			plt.plot(sparsities, fps, label=file.split("\\")[-3])
		else: 
			plt.plot(sparsities, fms, label=file.split("\\")[-3])
	
	plt.xlim(0, 0.2)
	plt.xlabel("Sparsity")
	plt.ylabel(metric.upper())
	plt.title(f"{metric.upper()} vs Edge Sparsity")
	plt.legend()
	plt.show()

```
